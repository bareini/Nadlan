```python
def expand_more_text(element, tag_name="span"):
    """
    This method extends the full text, in case this element exits

    :param element: the comment element
    :param str tag_name: the tag name to find
    :return: the comment element, after being expended
    """
    more_text = element.find_elements_by_tag_name(tag_name)
    # wait(element, 20)
    time.sleep(0.1)
    if len(more_text) > 0:
        try:
            more_text[0].click()  # click to get full text
        except ElementClickInterceptedException:    # fix the case in which the clickable object moved
            webdriver.ActionChains(browser).move_to_element(more_text[0]).click(more_text[0]).perform()
    return element
```

```python
def get_elements(element, tag_name, attr_name, attr_value, func=None) -> list:
    """
    This method find the relevant element in the page, using tag attribute and names

    :param element: the element to be searched in
    :param str tag_name: the tag name of the elements with the information we need
    :param str attr_name: the attribute of interest in the element
    :param str attr_value: the value of above attribute
    :param func: in case we want to use function on the returned element (e.g. expand more text)
    :return: list of relevant elements
    """
    relevant_elm = []
    for elm in element.find_elements_by_tag_name(tag_name):
        temp_elm = elm.get_attribute(attr_name)
        if temp_elm is not None and attr_value in temp_elm:
            # todo: check maybe not need to save in temp_element, as it change the element itself
            temp_elm = elm  # to make sure we don't change anything if not needed
            if func is not None:
                temp_elm = func(elm)  # expend the comment, if needed
            relevant_elm.append(temp_elm)
    return relevant_elm
```
